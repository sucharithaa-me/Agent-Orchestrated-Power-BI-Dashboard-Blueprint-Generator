```python
# ==========================================================
# IMPORTS
# ==========================================================

import pandas as pd
from pprint import pprint

from pandas.api.types import (
    is_numeric_dtype,
    is_datetime64_any_dtype,
    is_string_dtype
)

from google.colab import files
```

```python
# ==========================================================
# LOAD DATASET
# ==========================================================

uploaded = files.upload()

filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("Shape :", df.shape)

display(df.head())
```

```python
# ==========================================================
# BASIC DATASET INFORMATION
# ==========================================================

print("\nDataset Info")
df.info()

print("\nStatistics")
display(df.describe())

print("\nMissing Values")
display(df.isnull().sum())

print("\nDuplicate Rows")
print(df.duplicated().sum())

print("\nUnique Values")
display(df.nunique())
```

```python
# ==========================================================
# COLUMN CLASSIFICATION
# ==========================================================

column_classification = {}

total_rows = len(df)

for col in df.columns:

    tags = []

    unique_count = df[col].nunique()

    unique_ratio = unique_count / total_rows

    col_name = col.lower()

    # -----------------------------
    # DATE
    # -----------------------------

    if (
        is_datetime64_any_dtype(df[col])
        or any(word in col_name for word in
               ["date","time","year","month"])
    ):
        tags.append("Date")

    # -----------------------------
    # IDENTIFIER
    # -----------------------------

    if (
        any(word in col_name for word in
            ["id","code","number","no"])
        or unique_ratio > 0.95
    ):
        tags.append("Identifier")

    # -----------------------------
    # NUMERIC
    # -----------------------------

    if is_numeric_dtype(df[col]):

        tags.append("Numeric")

        if "Identifier" not in tags:
            tags.append("Measure")

        if unique_count <= 30:
            tags.append("Category")

    # -----------------------------
    # TEXT
    # -----------------------------

    elif is_string_dtype(df[col]):

        tags.append("Text")

        if unique_count <= 30:
            tags.append("Category")

    column_classification[col] = tags

print("\nCOLUMN CLASSIFICATION\n")

for col, tags in column_classification.items():
    print(f"{col:<35} -> {tags}")
```

```python
# ==========================================================
# METADATA
# ==========================================================

metadata = {

    "rows": len(df),

    "columns": len(df.columns),

    "duplicate_rows": int(df.duplicated().sum()),

    "missing_values": df.isnull().sum().to_dict(),

    "column_classification": column_classification

}

print("\nMETADATA\n")

pprint(metadata)
```

**PROMPTS**

```python
# ==========================================================
# DATASET CONTEXT
# ==========================================================

def build_dataset_context(metadata):

    context = f"""
DATASET SUMMARY
----------------
Rows            : {metadata["rows"]}
Columns         : {metadata["columns"]}
Duplicate Rows  : {metadata["duplicate_rows"]}

COLUMN CLASSIFICATION
----------------------
"""

    for column, tags in metadata["column_classification"].items():
        context += f"{column:<35}: {', '.join(tags)}\n"

    return context
```

```python
# ==========================================================
# DOMAIN PROMPT
# ==========================================================

def build_domain_prompt(metadata):

    prompt = "YOU ARE A SENIOR DATA ANALYST.\n\n"

    prompt += build_dataset_context(metadata)

    prompt += """

YOUR TASK

1. Identify the Business Domain.

2. Explain what the dataset is about.

3. Mention the Business Objective.

Keep your answer under 150 words.
"""

    return prompt
```

```python
# ==========================================================
# KPI PROMPT
# ==========================================================

def build_kpi_prompt(metadata, domain_response):

    prompt = "YOU ARE A BUSINESS KPI EXPERT.\n\n"

    prompt += build_dataset_context(metadata)

    prompt += f"""

BUSINESS DOMAIN
----------------
{domain_response}

YOUR TASK

Recommend the Top 10 KPIs.

For every KPI mention:

• KPI Name

• Why Important

• Formula (if applicable)

Return as a table.
"""

    return prompt
```

```python
# ==========================================================
# GENERIC AGENT
# ==========================================================

class Agent:

    def __init__(self,
                 name,
                 prompt_builder,
                 client):

        self.name = name
        self.prompt_builder = prompt_builder
        self.client = client

    def think(self, *args):

        print(f"\n{'='*60}")
        print(f"Running {self.name}")
        print(f"{'='*60}")

        prompt = self.prompt_builder(*args)

        response = self.client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )

        return response.text
```

```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
class AgentState(TypedDict):

    metadata: dict

    domain_response: str
    kpi_response: str
    visualization_response: str
    dashboard_response: str

    story_response: str
    insight_response: str
    quality_response: str
    dax_response: str
```

```python
def domain_node(state: AgentState):

    response = domain_agent.think(
        state["metadata"]
    )

    return {

        "metadata": state["metadata"],

        "domain_response": response,

        "kpi_response": "",

        "visualization_response": "",

        "dashboard_response": ""

    }
```

```python
def kpi_node(state: AgentState):

    response = kpi_agent.think(

        state["metadata"],

        state["domain_response"]

    )

    return {

        "metadata": state["metadata"],

        "domain_response": state["domain_response"],

        "kpi_response": response,

        "visualization_response": "",

        "dashboard_response": ""

    }
```

```python
def visualization_node(state: AgentState):

    response = visualization_agent.think(

        state["metadata"],

        state["domain_response"],

        state["kpi_response"]

    )

    return {

        "metadata": state["metadata"],

        "domain_response": state["domain_response"],

        "kpi_response": state["kpi_response"],

        "visualization_response": response,

        "dashboard_response": ""

    }
```

```python
def dashboard_node(state: AgentState):

    response = dashboard_agent.think(

        state["metadata"],

        state["domain_response"],

        state["kpi_response"],

        state["visualization_response"]

    )

    return {

        "metadata": state["metadata"],

        "domain_response": state["domain_response"],

        "kpi_response": state["kpi_response"],

        "visualization_response": state["visualization_response"],

        "dashboard_response": response

    }
```

```python
def story_node(state: AgentState):

    response = story_agent.think(
        state["metadata"]
    )

    return {
        **state,
        "story_response": response
    }
```

```python
def insight_node(state: AgentState):

    response = insight_agent.think(
        state["metadata"]
    )

    return {
        **state,
        "insight_response": response
    }
```

```python
def quality_node(state: AgentState):

    response = quality_agent.think(
        state["metadata"]
    )

    return {
        **state,
        "quality_response": response
    }
```

```python
def dax_node(state: AgentState):

    response = dax_agent.think(
        state["metadata"]
    )

    return {
        **state,
        "dax_response": response
    }
```

```python
graph = StateGraph(AgentState)
```

```python
graph.add_node("domain", domain_node)
graph.add_node("kpi", kpi_node)
graph.add_node("visualization", visualization_node)
graph.add_node("dashboard", dashboard_node)
```

```python
graph.add_node("story", story_node)
graph.add_node("insight", insight_node)
graph.add_node("quality", quality_node)
graph.add_node("dax", dax_node)
```

```python
graph.add_edge(START, "domain")

graph.add_edge("domain", "kpi")

graph.add_edge("kpi", "visualization")

graph.add_edge("visualization", "dashboard")

graph.add_edge("dashboard", "story")
graph.add_edge("story", "insight")
graph.add_edge("insight", "quality")
graph.add_edge("quality", "dax")
graph.add_edge("dax", END)
```

```python
app = graph.compile()
```

```python
from IPython.display import Image, display

display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({

    "metadata": metadata,

    "domain_response": "",

    "kpi_response": "",

    "visualization_response": "",

    "dashboard_response": "",

    "story_response": "",

    "insight_response": "",

    "quality_response": "",

    "dax_response": ""

})
```
